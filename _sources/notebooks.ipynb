```python
# Run the pip install command below if you don't already have the library
# !pip install git+https://github.com/JustAnotherArchivist/snscrape.git

# Run the below command if you don't already have Pandas
# !pip install pandas

# Imports
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
# Setting variables to be used below
maxTweets = 1000

# Creating list to append tweet data to
tweets_list1 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('Ridwan Kamil since:2022-08-01 until:2023-02-20').get_items()):
    if i>maxTweets:
        break
    tweets_list1.append([tweet.user.username, tweet.content, tweet.user.location])
```

```python
# Creating a dataframe from the tweets list above
tweets_df1 = pd.DataFrame(tweets_list1, columns=['Akun Tweet', 'Tweet', 'Location'])

# Display first 5 entries from dataframe
tweets_df1
```

```python
tweets_df1.to_csv('twitter.csv', index=False)
```
